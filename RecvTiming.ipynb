```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
%matplotlib notebook
```

```python
world_rank = 2
data = np.genfromtxt("build/out/recv_times_{}.csv".format(world_rank), delimiter=",", skip_header=1)
```

```python
n = data[:, 0]
dt = 1000*data[:, 1]
```

```python
plt.figure()
plt.loglog(n, dt, '+')
plt.show()
```

```python
n_val, n_count = np.unique(n, return_counts=True)


plt.figure()
plt.plot(n_val, n_count, '+-')
plt.show()
```
